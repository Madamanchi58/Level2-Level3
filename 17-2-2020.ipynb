```python
from sklearn.cluster import KMeans
l=[]
for i in range(1,11):
    kmeans1=KMeans(n_clusters=i,init='k-means++',random_state=1)
    kmeans1.fit(data_standard)
    l.append(kmeans1.inertia_) # inertia_ : Sum of squared distance to their closest cluster center
```

```python
y_pred=kmeans1.fit_predict(data_standard)
```
