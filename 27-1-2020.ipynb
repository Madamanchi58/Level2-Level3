### Scatter Plot

```python
x=[1,2,3,4,5]
y=[2,5,6,7,8]
plt.scatter(x,y,color='b')
plt.title("Scatter Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

### Area Plot

```python
days=[1,2,3,4]
working=[8,8.5,9,10]
sleeping=[8,8,11,12]
plt.stackplot(days,working,sleeping,colors=['purple','pink'],labels=["working","sleeping"])
plt.legend()
plt.title('Area Plot')
plt.xlabel('days')
plt.ylabel('sleeping@working')
plt.show()
```

### Pie Chart

```python
days=[1,2,3,4]
working=[8,8.5,9,10]
sleeping=[8,8,11,12]
slices=[4,10,12]
variable=['days','working','sleeping']
plt.pie(slices,labels=variable,colors=['g','r','b'])
plt.show()
```

```python
days=[1,2,3,4]
working=[8,8.5,9,10]
sleeping=[8,8,11,12]
slices=[4,10,12]
variable=['days','working','sleeping']
plt.pie(slices,labels=variable,colors=['g','r','b'],startangle=90)
plt.show()
```

```python
days=[1,2,3,4]
working=[8,8.5,9,10]
sleeping=[8,8,11,12]
slices=[4,10,12]
variable=['days','working','sleeping']
plt.pie(slices,labels=variable,colors=['g','r','b'],startangle=90,autopct='%1.2f%%')
plt.show()
```
